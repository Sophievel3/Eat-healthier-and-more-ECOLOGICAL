```python
                                              ## R É S U M É ##

##Stratégie de remplissage de données manquantes:

##- Le 'code' est toujours renseignés (c'est la base de "Open Food Facts").

##- Le 'countries_en' aussi (ie. on l'a choisit ainsi car notre étude se limite à la France).

##- Le 'product_name' n'est pas vraiment important on l'utilisera pour l'affichage final, mais pas vraiment pour le 
##score.

##- Le 'General-carbon-footprint', s'obtient en fusionnant le colonnes 'carbon-footprint_100g' et 
##'carbon-footprint-from-meat-or-fish_100g'.

##- les 'ingredients_that_may_be_from_palm_oil_n' et 'ingredients_from_palm_oil_n', on ne peut pas les inventer non 
##plus, ils sont utilisés pour remplir des données manquantes du 'General-carbon-footprint'. En fait les données 
##scientifiques expertes ont démontré que, par hectare, la quantité de carbone émise due à la culture de l'huile de 
##palme est à peu près de 170 tonnes, avec un rendement moyen mondial de 3.7 tonnes à l'hectare, soit 45g par g d'huile 
##de palme.
##Ainsi on a complété les lignes non renseignées du 'General-carbon-footprint' en ajoutant 45g par ingédient (ceci car
##on prend toujours le pire scénario car on veut inciter à des produit plus écologiques, et donc qui de préférence n'ont
##pas d'ingrédients issus de l'huile de palme), trouvant ainsi le 'Final-carbon-footprint'. Les derniers 'NaN' du
#Final-carbon-footprint' sont complétés par les médianes des 'Final-carbon-footprint' associés a leur 'main_category_en'.
#L'absence de 'main-categorie_en' est donc de nouveau contraignante.


##- Les 'nutriscore_grade', 'nutriscore_score', et 'nutrition-score-fr_100g', on ne peux pas les inventer. Ils sont 
##calculés par des formules établies.Du coup on commence par merger les lignes en doubles, afin de compléter quelques 
#lignes. Ensuite on fusionne les colonnes 'nutriscore_score' et 'nutrition-score-fr_100g' pour faire un seul indice 
##'nutriscore_final'. On complète les lignes manquantes du 'nutriscore_final' avec la médiane de leur catégorie 
#respective +1, pour ensuite le transformer en 'nutriscore_grade_final' avec les formules de Santé Publique France.
##Finalement on vérifie la cohérence entre 'nutriscore_grade' et 'nutriscore_grade_final' pour valider notre 
#'nutriscore_grade_final' et créer 'nutriscore_grade_Total' qui prend le pire score entre 'nutriscore_grade' et 
#'nutriscore_grade_final'.


##- les 'additives_n', on ne peut pas les inventer mais, quand ce n'est pas renseigné, on associé la médiane de 
#la 'main_category_en' à laquelle ils appartiennent, pour obtenir 'addtives_Total'. L'absence de 'main-categorie_en' 
#est donc de nouveau contraignante.

##- La 'main_category_en' on ne peut pas se mettre à la renseignée produit par produit, du coup les lignes qui ne l'ont 
## pas et qui n'ont pas d' 'additives_n' renseignés sont éliminées.

##Les dernières lignes avec des 'NaN' dans les colonnes 'additives_n' et 'Final-carbon-footprint' sont éliminées à la 
##toute fin.

## ATTENTION on choisit les médianes car elle représentent mieux la tendance.
## ATTENTION on ne complétera pas les 'nutriscore_grade_final' qui restent avec des 'NaN' en utilisant les médianes de
#leur 'main_category_en'. Contrairement aux 'additives_Total' et au 'Final-carbon-footprint', qui concernent la présence
#d'un élément, le 'nutriscore_grade_final' est lié à la quantité présente d'un élément. Ce qui fait que prendre la 
#médiane serait aberrant car il peut le changer de façon radicale. Pour cela on préfère constituer une nouvelle 
#catégorie "n" pour ces données là.
```
